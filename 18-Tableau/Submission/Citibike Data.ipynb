```python
import pandas as pd
import numpy as np
```

```python
months = ["1901", "1902", "1903", "1904", "1905", "1906", "2001", "2002", "2003", "2004", "2005", "2006"]
main_df = pd.DataFrame()

for month in months:
    print(month)
    filepath = f"Data/data/20{month}-citibike-tripdata.csv"
    df = pd.read_csv(filepath, low_memory=False)
    df = df.sample(frac=0.015, random_state=42)
    main_df = main_df.append(df)
    
main_df = main_df.reset_index(drop=True)
```

```python
main_df.to_csv("19Q1vs20Q1_citibike_rides_sample.csv", index=False)
```
